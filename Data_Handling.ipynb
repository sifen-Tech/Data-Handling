```python
import pandas as pd

titanic = pd.read_csv("dataset-2.csv")
titanic.head()
```

```python
print(titanic.head())
print(titanic.info())
print(titanic.describe())
```

```python
# Fill missing Age with median (safe way)
titanic["Age"] = titanic["Age"].fillna(titanic["Age"].median())

# Fill missing Embarked with mode (safe way)
titanic["Embarked"] = titanic["Embarked"].fillna(titanic["Embarked"].mode()[0])

# Drop Cabin column ONLY if it exists
if "Cabin" in titanic.columns:
    titanic = titanic.drop(columns="Cabin")

# Remove duplicate rows
titanic = titanic.drop_duplicates()
```

```python
# Survival rate by gender
survival_gender = titanic.groupby("Sex")["Survived"].mean()
print("Survival by Gender:\n", survival_gender)

# Survival rate by class
survival_class = titanic.groupby("Pclass")["Survived"].mean()
print("\nSurvival by Class:\n", survival_class)

# Average age per class
avg_age_class = titanic.groupby("Pclass")["Age"].mean()
print("\nAverage Age per Class:\n", avg_age_class)

# Create age groups
titanic["AgeGroup"] = pd.cut(titanic["Age"], bins=[0, 12, 18, 60, 100],
                             labels=["Child", "Teen", "Adult", "Senior"])

# Survival rate by age group
survival_age = titanic.groupby("AgeGroup")["Survived"].mean()
print("\nSurvival by Age Group:\n", survival_age)
```

```python
# Female passengers who survived
female_survived = titanic[(titanic["Sex"] == "female") & (titanic["Survived"] == 1)]

# Children who survived
children_survived = titanic[(titanic["Age"] < 12) & (titanic["Survived"] == 1)]

# First class passengers with high survival probability
first_class_survived = titanic[(titanic["Pclass"] == 1) & (titanic["Survived"] == 1)]

print(female_survived.head())
print(children_survived.head())
print(first_class_survived.head())
```

        Insights
1. Who was more likely to survive?
Females had a much higher survival rate than males.
2. Did class affect survival?
Yes. 1st class passengers had the highest survival rate.
3rd class had the lowest survival rate.
3. Were children prioritized?
Yes. Children had relatively higher survival rates, showing priority in rescue.
4. What combination had highest survival rate?
Female + 1st Class + Child
This group had the highest chance of survival.
